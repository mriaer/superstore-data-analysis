# **STEP 2: DATA CLEANING**

---





> Data cleaning is the process of fixing or removing incorrect, corrupted, incorrectly formatted, duplicate, or incomplete data within a dataset. When combining multiple data sources, there are many opportunities for data to be duplicated or mislabeled. If data is incorrect, outcomes and algorithms are unreliable, even though they may look correct. There is no one absolute way to prescribe the exact steps in the data cleaning process because the processes will vary from dataset to dataset. But it is crucial to establish a template for your data cleaning process so you know you are doing it the right way every time.



```python
# Checking missing value
df_superstore.isnull().values.any()
```

```python
# Delete the data with missing values and check again
df_clean = df_superstore.dropna()
df_clean.isnull().values.any()
```

```python
# Check for duplicates in the dataset
df_duplicates = df_superstore.duplicated(keep=False)
df_duplicates.sum()
```

```python
# Dropping duplicate rows and check again
df_superstore.drop_duplicates(inplace=True)
df_superstore.shape
```

```python
# Return data type of each columns
df_superstore.dtypes
```

```python
# Checking summary statistics of non-numerical columns
df_superstore.describe(include=object)
```

```python
# Adding some time series columns
df_superstore['Year'] = pd.to_datetime(df_superstore['Order Date']).dt.year
df_superstore['Month'] = pd.to_datetime(df_superstore['Order Date']).dt.month_name()
df_superstore['Lead Time'] = pd.to_datetime(df_superstore['Ship Date']) - pd.to_datetime(df_superstore['Order Date'])
df_superstore['Lead Time'] = df_superstore['Lead Time'].dt.days
df_superstore.info()
```

# **STEP 3: EXPLORATORY DATA ANALYSIS**

---





> Exploratory Data Analysis (EDA) is a process of conducting preliminary inspections/investigations on data to discover patterns and anomalies, as well as to test hypotheses and to check assumptions with the help of summary statistics and graphical visualization representations.



```python
# Pivot table for best selling by category
pivot_1 = df_superstore.groupby('Category')['Quantity'].sum().sort_values(ascending=False)
pivot_1
```

```python
# 1st table pivot chart
plt.figure(figsize=(10, 6))
pivot_1.plot(kind='bar', color=['red', 'blue', 'green'])
plt.title('Best Selling by Category')
plt.xlabel('Category')
plt.ylabel('Total Quantity Sold')
plt.xticks(rotation=0)
plt.tight_layout()
plt.grid(axis='y')
plt.show()
```

****

```python
# Pivot table for most profitable by category
pivot_2 = df_superstore.groupby('Category')['Profit'].sum().sort_values(ascending=False)
pivot_2
```

```python
# 2nd table pivot chart
plt.figure(figsize=(10, 6))
pivot_2.plot(kind='bar', color=['green', 'red', 'blue'])
plt.title('Most Profitable Categories')
plt.xlabel('Category')
plt.ylabel('Total Profit')
plt.xticks(rotation=0)
plt.tight_layout()
plt.grid(axis='y')
plt.show()
```

```python
# Pivot table for top 10 product by sales
pivot_3 = df_superstore.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)
pivot_3
```

```python
# 3rd table pivot chart
plt.figure(figsize=(10, 6))
pivot_3.plot(kind='bar', color=['orange', 'purple', 'brown', 'pink', 'gray', 'olive', 'cyan', 'black', 'yellow', 'lightgreen'])
plt.title('Top 10 Selling Products by Sales')
plt.xlabel('Product Name')
plt.ylabel('Total Sales')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.show()
```

```python
# Pivot table for top 10 city by sales
pivot_4 = df_superstore.groupby('City')['Sales'].sum().sort_values(ascending=False).head(10)
pivot_4
```

```python
# 4th table pivot chart
plt.figure(figsize=(10, 6))
pivot_4.plot(kind='bar', color=['silver', 'lightcoral', 'maroon', 'mistyrose', 'tomato', 'chocolate', 'yellowgreen', 'indigo', 'crimson', 'dimgray'])
plt.title('Top 10 Cities by Sales')
plt.xlabel('City')
plt.ylabel('Total Sales')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.grid(axis='y')
plt.show()
```

```python
# Pivot table for monthly sales trend
pivot_5 = df_superstore.groupby(['Year', 'Month'])['Sales'].sum().reset_index()
pivot_5['Month'] = pd.Categorical(pivot_5['Month'], categories=['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'], ordered=True)
pivot_5 = pivot_5.sort_values(['Year', 'Month']).set_index(['Year', 'Month'])['Sales'].unstack()
pivot_5
```

```python
# 5th table pivot chart
plt.figure(figsize=(10, 6))
for year in pivot_5.index:
    plt.plot(pivot_5.columns, pivot_5.loc[year], label=year, marker='o')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.legend()
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.grid(axis='y')
plt.show()
```

```python
# Pivot table for average lead time by shipping mode and category
pivot_6 = df_superstore.groupby(['Ship Mode', 'Category'])['Lead Time'].mean().unstack()
pivot_6
```

```python
# 6th table pivot chart
plt.figure(figsize=(10, 6))
pivot_6.plot(kind='bar')
plt.title('Average Lead Time by Shipping Mode and Category')
plt.xlabel('Shipping Mode')
plt.ylabel('Average Lead Time (Days)')
plt.xticks(rotation=0)
plt.tight_layout()
plt.grid(axis='y')
plt.show()
```

```python
# Pivot table for average discount by region
pivot_7 = df_superstore.groupby('Sub-Category')['Discount'].mean()
pivot_7
```

```python
# 7th table pivot chart
plt.figure(figsize=(10, 6))
pivot_7.plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Average Discount by Sub-Category')
plt.axis('equal')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

```python
# Pivot table for total profit by region
pivot_8 = df_superstore.groupby('Region')['Profit'].sum()
pivot_8
```

```python
# 8th table pivot chart
plt.figure(figsize=(10, 6))
pivot_8.plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Total Profit by Region')
plt.axis('equal')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

```python
# Pivot table for top 10 Customer by Quantity
pivot_9 = df_superstore.groupby('Customer Name')['Quantity'].sum().sort_values(ascending=False).head(10)
pivot_9
```

```python
# 9th pivot table chart
plt.figure(figsize=(10, 6))
pivot_9.plot(kind='barh', color=['b', 'g', 'r', 'c', 'm', 'y', 'k', '#ff9999', '#66b3ff', '#99ff99'])
plt.title('Top 10 Customers by Quantity')
plt.xlabel('Customer Name')
plt.ylabel('Total Quantity')
plt.xticks(rotation=0)
plt.tight_layout()
plt.grid(axis='both')
plt.show()
```

```python
# Pivot table for average Sales per Order by Customer
pivot_10 = df_superstore.groupby('Customer Name')['Sales'].mean().sort_values(ascending=False).head(5)
pivot_10
```

```python
# 10th pivot table chart
plt.figure(figsize=(10, 6))
pivot_10.plot(kind='barh', color=['r', 'g', 'b', 'black', 'purple'])
plt.title('Average Sales per Order by Customer')
plt.xlabel('Customer Name')
plt.ylabel('Average Sales')
plt.xticks(rotation=0)
plt.tight_layout()
plt.grid(axis='both')
plt.show
```
